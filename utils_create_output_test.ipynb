```python
import pandas as pd
```

```python
test_after_match_1 = pd.read_csv("/content/test_after_match_1.csv")
test_after_match_2 = pd.read_csv("/content/test_after_match_2.csv")
test_after_match_4 = pd.read_csv("/content/test_after_match_4.csv")
summ_after_match = pd.read_csv("/content/summ_after_match.csv")

test_during_match_1 = pd.read_csv("/content/test_during_match_1.csv")
test_during_match_3 = pd.read_csv("/content/test_during_match_3.csv")
summ_during_match = pd.read_csv("/content/summ_during_match.csv")
```

```python
def concatenate_content(messages):
  return ' '.join([str(message).strip() for message in messages])

summ_after_match = summ_after_match[summ_after_match['relevant_part'] == 1]
summ_after_match.drop(['relevant_part', 'type'], axis=1, inplace=True)


grouping_columns = ['match_id', 'news_id', 'name', 'team1', 'team2', 'news_name_body']
summ_after_match = summ_after_match.sort_values(['match_id', 'news_id', 'minute'], ascending = True).groupby(grouping_columns)['content'].apply(concatenate_content).reset_index()
summ_after_match
```

```python
output_test_after_match_4 = summ_after_match.merge(test_after_match_4[['match_id', 'news_id']].drop_duplicates(), how='inner', left_on=['match_id', 'news_id'], right_on=['match_id', 'news_id'])
```

```python
def concatenate_content(messages):
  return ' '.join([str(message).strip() for message in messages])

summ_during_match = summ_during_match[summ_during_match['relevant_part'] == 1]
summ_during_match.drop(['relevant_part', 'type'], axis=1, inplace=True)


grouping_columns = ['match_id', 'news_id', 'name', 'team1', 'team2', 'news_name_body']
summ_during_match = summ_during_match.sort_values(['match_id', 'news_id', 'minute'], ascending = True).groupby(grouping_columns)['content'].apply(concatenate_content).reset_index()
summ_during_match
```

```python
output_test_during_match_3 = summ_during_match.merge(test_during_match_3[['match_id', 'news_id']].drop_duplicates(), how='inner', left_on=['match_id', 'news_id'], right_on=['match_id', 'news_id'])
```

```python
after_match_new_news_1, after_match_new_news_2, after_match_new_news_4 = [], [], []
during_match_new_news_1, during_match_new_news_3 = [], []

with open("/content/after_match_generated_predictions1.txt", 'r', encoding='utf-8') as f:
  for line in f.readlines():
    after_match_new_news_1.append(line.strip('\n'))

with open("/content/after_match_generated_predictions2.txt", 'r', encoding='utf-8') as f:
  for line in f.readlines():
    after_match_new_news_2.append(line.strip('\n'))

with open("/content/after_match_generated_predictions4.txt", 'r', encoding='utf-8') as f:
  for line in f.readlines():
    after_match_new_news_4.append(line.strip('\n'))


with open("/content/during_match_generated_predictions1.txt", 'r', encoding='utf-8') as f:
  for line in f.readlines():
    during_match_new_news_1.append(line.strip('\n'))

with open("/content/during_match_generated_predictions3.txt", 'r', encoding='utf-8') as f:
  for line in f.readlines():
    during_match_new_news_3.append(line.strip('\n'))
```

```python
test_after_match_1['SumRes_1'] = after_match_new_news_1
test_after_match_1.rename(columns={"content": "input_SumRes_1"}, inplace=True)
test_after_match_1['input_SumRes_2'] = test_after_match_2['content']
test_after_match_1['SumRes_2'] = after_match_new_news_2
```

```python
output_test_after_match_4.rename(columns={"content": "input_SumRes_main"}, inplace=True)
output_test_after_match_4['SumRes_main'] = after_match_new_news_4
```

```python
res_after_match = test_after_match_1.merge(output_test_after_match_4[['match_id', 'news_id', 'SumRes_main', 'input_SumRes_main']], how='left', left_on=['match_id', 'news_id'], right_on=['match_id', 'news_id'])
```

```python
test_during_match_1['SumDur_1'] = during_match_new_news_1
test_during_match_1.rename(columns={"content": "input_SumDur_1"}, inplace=True)
```

```python
output_test_during_match_3.rename(columns={"content": "input_SumDur_main"}, inplace=True)
output_test_during_match_3['SumDur_main'] = during_match_new_news_3
```

```python
res_during_match = test_during_match_1.merge(output_test_during_match_3[['match_id', 'news_id', 'SumDur_main', 'input_SumDur_main']], how='left', left_on=['match_id', 'news_id'], right_on=['match_id', 'news_id'])
```

```python
res_after_match.drop(['news_name_body'], axis=1, inplace=True)
res_during_match.drop(['news_name_body'], axis=1, inplace=True)
```

```python
res_after_match = res_after_match[['match_id', 'news_id', 'name', 'team1', 'team2', 'news_name', 'news_body', 'input_SumRes_1', 'SumRes_1', 'input_SumRes_2', 'SumRes_2', 'input_SumRes_main', 'SumRes_main']]
res_during_match = res_during_match[['match_id', 'news_id', 'name', 'team1', 'team2', 'news_name', 'news_body', 'input_SumDur_1', 'SumDur_1', 'input_SumDur_main', 'SumDur_main']]
```

```python
res_after_match.to_csv('/content/res_after_match.csv', index=False, encoding='utf-8')
res_during_match.to_csv('/content/res_during_match.csv', index=False, encoding='utf-8')
```
