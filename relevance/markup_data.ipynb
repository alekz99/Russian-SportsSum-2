```python
import pandas as pd
import numpy as np
import random
from pprint import pprint
from IPython.display import clear_output 

data = pd.read_csv("/content/result_news.csv")
data
```

```python
data['length_news_name'] = data['name'].apply(lambda x: len(x.split()))
data['length_body'] = data['body'].apply(lambda x: len(x.split()))
data[['length_news_name', 'length_body']].describe()
```

```python
data.drop(['length_news_name', 'length_body'], axis=1, inplace=True)
```

```python
good_news = []
bad_news = []
indexes = data.index.tolist()
random.shuffle(indexes)

for index in indexes:
  print('len(good_news)=', len(good_news))
  if len(good_news) == 750:
    break
  row = data.loc[index]
  print(row['name'])
  print('--------------------')
  pprint(row['body'], width=150)
  
  label = int(input())
  if label == 1:
    good_news.append(index)
  elif label == 0:
    bad_news.append(index)

  clear_output(True)
```

```python
good_news
```

```python
bad_news
```

```python
good_train = good_news[:500]
good_eval = good_news[500:]

bad_train = bad_news[:550]
bad_eval = bad_news[550:]
```

```python
data_good_train_relevance = data.loc[good_train]
data_good_train_relevance['relevance'] = 1

data_bad_train_relevance = data.loc[bad_train]
data_bad_train_relevance['relevance'] = 0

train = pd.concat([data_good_train_relevance, data_bad_train_relevance])
train = train.sample(frac=1, random_state=42)
train
```

```python
# добавочные новости для train. пул новостей при активном обучении
indexes = data.index.tolist()
random.shuffle(indexes)

add_news = []

for index in indexes:
  if index not in good_news and index not in bad_news:
    add_news.append(index)
```

```python
data_add_train_relevance = data.loc[add_news]
data_add_train_relevance['relevance'] = 0


pool_train = pd.concat([train, data_add_train_relevance])
pool_train
```

```python
data_good_eval_relevance = data.loc[good_eval]
data_good_eval_relevance['relevance'] = 1

data_bad_eval_relevance = data.loc[bad_eval]
data_bad_eval_relevance['relevance'] = 0

eval = pd.concat([data_good_eval_relevance, data_bad_eval_relevance])
eval = eval.sample(frac=1, random_state=42)
eval
```

```python
pool_train['name_body'] = pool_train['name'] + ". " + pool_train['body']
eval['name_body'] = eval['name'] + ". " + eval['body']
```

```python
pool_train.to_csv('train.csv', index=False, encoding='utf-8')
```
